## Panorama Stitching

````python
import cv2 # Important Library
import matplotlib.pyplot as plt # For showing the image
import numpy as np # Used to manipulate image as arrays
````

We load our input images.

````python
img1 = cv2.imread("/Users/jeslyvarghese/Workspace/Extra/opencv_extra/testdata/stitching/boat1.jpg", 0)
img2 = cv2.imread("/Users/jeslyvarghese/Workspace/Extra/opencv_extra/testdata/stitching/boat2.jpg", 0)
````

````python
plt.imshow(img1, cmap = "gray")
````

````python
plt.imshow(img2, cmap = "gray")
````

Detect keypoints or features in the image. There are various algos SIFT, SURF (non-free) 

````python
sift = cv2.xfeatures2d.SIFT_create()
````

Get keypoints and keypoint descriptors

````python
kp1, des1 = sift.detectAndCompute(img1, None)
````

````python
kp2, des2 = sift.detectAndCompute(img2, None)
````

````python
kp_img1 = np.zeros([img1.shape[0], img1.shape[1], 3], dtype=np.uint8)
````

````python
cv2.drawKeypoints(img1, kp1, kp_img1, color=(0, 255, 0), flags=0)
````

````python
plt.imshow(kp_img1)
````

````python
kp_img2 = np.zeros([img2.shape[0], img2.shape[1], 3], dtype=np.uint8)
````

````python
cv2.drawKeypoints(img2, kp2, kp_img2)
````

````python
plt.imshow(kp_img2)
````

We match between keypoints or features with their descriptors
- using FLANN matcher here, there exist a BF-Matcher

````python
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks = 50)
````

````python
flann = cv2.FlannBasedMatcher(index_params, search_params)
````

````python
matches = flann.knnMatch(des1, des2,k=2)
````

Find good matches (D.Lowe Paper). 

````python
good = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:
        good.append(m)
````

````python
MIN_MATCH_COUNT = 10
if len(good) > MIN_MATCH_COUNT: # threshold the number of matches needed to find H matrix
    src_pts = np.float32([kp2[m.trainIdx].pt for m in good]).reshape(-1, 1, 2) # create points to map from
    dst_pts = np.float32([kp1[m.queryIdx].pt for m in good]).reshape(-1, 1, 2) # create points to map to
    M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC, 5.0) # find homography matrix H that will map src->dst
    matchesMask = mask.ravel().tolist()
    h, w = img2.shape
    pts = np.float32([[0, 0], [0, h-1], [w-1, h-1], [w-1, 0]]).reshape(-1, 1, 2)
    dst = cv2.perspectiveTransform(pts, M) # apply homography matrix to perform the transform
    img2 = cv2.polylines(img2, [np.int32(dst)], True, 255, 3, cv2.LINE_AA)
else:
    print("Not enough matches found")
    matchesMask = None
````

````python
plt.imshow(img2)
````

Obtain the width and height of both the images

````python
r1, c1 = img1.shape[:2]
r2, c2 = img2.shape[:2]
````

We create an array of arrays,
||0 0| |0 w1| |h1 w1| |h1 0||
With reshape (-1, 1, 2) means set dimension of the array, -1 for unknown.
Here we want to have n number of rows, with 1 column each which can have 2 columns.

````python
l_pts1 = np.float32([[0, 0], [0, r1], [c1, r1], [c1, 0]]).reshape(-1, 1, 2)
display(l_pts1)
````

We combine dst array with the l_pts1 array.

````python
lop = np.concatenate((l_pts1, dst), axis=0)
````

````python
display(lop)
````

We find min and max x,y values. ravel() function just unflattens an array.

````python
[x_min, y_min] = np.int32(lop.min(axis=0).ravel() - 0.5)
[x_max, y_max] = np.int32(lop.max(axis=0).ravel() + 0.5)
display(x_min, y_min, x_max, y_max)
````

````python
translation_dist = [-x_min, -y_min]
display(translation_dist)
````

We create a translation matrix using the min translation in x and y direction.
Translation matrix are of the form
```
1 0 tx
0 1 ty
0 0 1
```

````python
H_translation = np.array([[1, 0, translation_dist[0]], [0, 1, translation_dist[1]], [0,0,1]])
display(H_translation)
````

We project the homography matrix M on to translation Matrix H_translation.
We do this inorder to adjust the homography matrix to include the entire part of image 2.

````python
H = H_translation.dot(M)
````

````python
display(H)
````

````python
output_img = cv2.warpPerspective(img2, H, (x_max - x_min, y_max - y_min))
````

````python
plt.imshow(output_img)
````

````python
output_img[translation_dist[1]:r1+translation_dist[1],translation_dist[0]:c1+translation_dist[0]] = img1
````

````python
plt.imshow(output_img)
````
